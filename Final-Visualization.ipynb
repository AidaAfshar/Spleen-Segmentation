```python
import os
import numpy as np
import glob
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
root_dir = 'test_predictions'
data_dir1 = os.path.join(root_dir,"best_model")
data_dir2 = os.path.join(root_dir,"avg_model")
data_dir3 = os.path.join(root_dir,"weighted_avg_model")
```

```python
best_model_labels = sorted(
    glob.glob(os.path.join(data_dir1, "labelsTs", "*.nii.gz")))
avg_model_labels = sorted(
    glob.glob(os.path.join(data_dir2, "labelsTs", "*.nii.gz")))
weighted_avg_model_labels = sorted(
    glob.glob(os.path.join(data_dir3, "labelsTs", "*.nii.gz")))
```

```python
best_model_labels
```

```python
root_dir = 'MONAI_DATA_DIRECTORY'
data_dir = os.path.join(root_dir,"3D_Spleen")
print(data_dir)
```

```python
test_images = sorted(
    glob.glob(os.path.join(data_dir, "imagesTs", "*.nii.gz")))
data_dicts = [
    {"image": image_name}
    for image_name in zip(test_images)
]
test_images = data_dicts[:]
```

```python
test_images
```

```python
images = []
best_model = []
avg_model = []
weighted_avg_model = []
```

```python
for i in range(20):
  images += [nib.load(data_dicts[i]["image"][0])]
  best_model += [nib.load(best_model_labels[i])]
  avg_model += [nib.load(avg_model_labels[i])]
  weighted_avg_model += [nib.load(weighted_avg_model_labels[i])]
```

```python
for i in range(20):
  images[i] = images[i].get_data()
  best_model[i] = best_model[i].get_data()
  avg_model[i] = avg_model[i].get_data()
  weighted_avg_model[i] = weighted_avg_model[i].get_data()
```

```python
images[0].shape
```

```python
for i in range(20):
      image = images[i]

      _,_,z = image.shape
      z = int(2*z/3)

      plt.figure("check", (24, 6))

      plt.subplot(1, 4, 1)
      plt.title(f"image {i}")
      plt.imshow(images[i][:, :, z], cmap="gray")

      plt.subplot(1, 4, 2)
      plt.title(f"best model {i}")
      plt.imshow(images[i][:, :, z], cmap="gray")
      plt.imshow(best_model[i][:, :, z],alpha=0.6,cmap="Reds")
      
      plt.subplot(1, 4, 3)
      plt.title(f"average {i}")
      plt.imshow(images[i][:, :, z], cmap="gray")
      plt.imshow(avg_model[i][:, :, z],alpha=0.6,cmap="Greens")
      
      plt.subplot(1, 4, 4)
      plt.title(f"weighted average {i}")
      plt.imshow(images[i][:, :, z], cmap="gray")
      plt.imshow(avg_model[i][:, :, z],alpha=0.6,cmap="Blues")
      
      
      plt.show()
```
